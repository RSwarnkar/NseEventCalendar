```python
import pandas as pd
import requests
import json
import datetime
```

```python
next_day = datetime.datetime.now() + datetime.timedelta(days=1)
next_day = next_day.strftime("%d-%m-%Y")
print(next_day)
```

```python
# List of URLs
url1 = "https://www.nseindia.com/"
url2 = "https://www.nseindia.com/companies-listing/corporate-filings-event-calendar"
url3 = "https://www.nseindia.com/api/event-calendar?index=equities&from_date="+next_day+"&to_date="+next_day

print(url3)

# Header
header1 = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"
}
```

```python
s = requests.Session()

s.headers.update(header1)
r1 = s.get(url2)

s.headers.update(header1)
r2 = s.get(url3, cookies=r1.cookies)
```

```python
if r2.status_code == 200:
    event_data = json.loads(r2.text)
```

```python
df = pd.DataFrame.from_dict(event_data, orient='columns')
```
